## 2. Функция plot для построения и оформления двумерных графиков

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Функция **plot** отображает обычные двумерные графики

```python
plt.plot([1, -2, 4, 1])
```

По оси **Y** отображаются значения из переданного списка, а по оси **X** - индекс, так работает график по умолчанию

Под капотом функции список меняет тип данных на numpy массив

Функция **plot** может принимать данные по оси **X**, так и по оси **Y** с помощью аргументов

```python
x = np.array([3, 5, 7, 9])
y = np.array([1, 15, 6, 10])
```

```python
plt.plot(x, y)
```

В **jupyter notebook** все вызываемые функции необходмо использовать в одной строке, тогда график будет построен правильно и последовательно

```python
x1 = np.array([i + 1 for i in x])
y1 = np.array([i + i for i in y])
```

Иногда необходимо наложить графики друг на друга, достаточно передать нужные аргументы, при это кол-во точек в массивах может быть разное

```python
plt.plot(x, y, x1, y1) # наложение графиков друг на друга
plt.grid() # отображение сетки
```

Ещё один способ построения такого же графика:

```python
plt.plot(x, y)
plt.plot(x1, y1)
plt.grid()
```

### Изменение цвета линий графиков

Цвета линий также меняются с помощью параметров

```python
plt.plot(x, y, 'b', x1, y1, 'r') # параметр является буквой
```

```python
plt.plot(x, y, 'b:', x1, y1, 'r-.') # параметр является буквой с типом линии
```

Можно явно указать параметр **color** для сразу всех графиков

```python
plt.plot(x, y, x1, y1, color='Orange')
```

Преимущество данного параметра, то что можно задавать разные цветовые схемы, типа RGB и т.д.

```python
plt.plot(x, y, x1, y1, color=(0.30, 0.20, 0.90, 0.5)) # 4 цифра это степень прозрачности
```

### Изменение маркеров точек у графиков

Существует возможность изменения маркеров у точек

```python
plt.plot(x, y, 'r-o', x1, y1, 'm:s')
```

Другой способ определения маркера – использование параметра **marker**:

```python
plt.plot(x, y, '-go', x1, y1, 's:m', marker='d')
```

Для задания цвета маркера, отличного от цвета линии, применяется параметр markerfacecolor:

```python
plt.plot(x, y, '-go', x1, y1, 's:m', marker='D', markerfacecolor='w', linewidth=3)
```

### Заливка областей графика

Наконец, можно делать заливку областей графика с помощью функции:

```python
fill_between(x,  y1,  y2=0,  where=None,  interpolate=False,  step=None,  *, data=None, **kwargs)
```

Основные параметры здесь следующие:

* x, y1 – массивы значений координат x и функции y1;  
* y2 – массив (или число) для второй кривой, до которой производится заливка;  
* where – массив булевых элементов, который определяет области для заливки.  

В самом простом случае эту функцию можно использовать так:

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
y = np.cos(x)
plt.plot(x, y)
plt.fill_between(x, y)
```

У нас получилась косинусоида с заливкой между значениями функции y и осью абсцисс y2 = 0. Если третьим параметром указать другое число, отличное от нуля, например, 0,5:

```python
plt.fill_between(x, y, 0.5)
```

А если дополнительно еще сделать ограничение на выбор заливаемого региона, когда y < 0:

```python
plt.fill_between(x, y, 0.5, where=(y < 0))
```

Также можно вызвать эту функцию два раза подряд

```python
plt.fill_between(x, y, where=(y < 0), color='r', alpha=0.5)
plt.fill_between(x, y, where=(y > 0), color='g', alpha=0.5)
```

Вот так можно с помощью функции plot() отображать графики в координатных осях и делать их простое оформление.
